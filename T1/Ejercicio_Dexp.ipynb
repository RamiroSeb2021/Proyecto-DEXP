Ahora vamos a mirar:

$$
\begin{cases}
    \mu_1 - \mu_2 = 0\\
    \mu_1 - \mu_2 = 0
\end{cases}
$$

Donde $\mu_1$ y $\mu_2$ son las poblaciones



*   $\mu_1$ = Halicta_oleracea
*   $\mu_2$ = Halicta_carduorum



```r
# Matrices de varianzas y covarianzas estimadas

S1 = cov(Halicta_oleracea[, -1])
S2 = cov(Halicta_carduorum[, -1])

n1 = dim(Halicta_oleracea)[1]
n2 = dim(Halicta_carduorum)[1]
# Media común
Sp = ((n1 - 1)*S1)+(n2 - 1)*S2/(n1+n2-2)
Sp
```

```r
# Estimación
k = 4
n = n2 + n1
factor_ = ((n*(n-k))/(k*(n-1)))
x_ = media_oleracea - media_carduorum
estadistico = factor_ * t(media_oleracea - media_carduorum)%*%solve(Sp)%*%(media_oleracea - media_carduorum)
```

Hay dos formas de hacer el cálculo de este contraste de hipotesis:



1.   Después de calcular el estadístico, compararlo con el quantil de la distribución de la $F$ y mirar si este varlor es menor o mayor que este, si es mayor, cae en la region de rechazo, si no la hipotesis nula no se rechaza
2.   Calcular el $p-value$ y mirar si es menor al $\alpha$,  entonces se rechazaía $H_0$



Forma 1

```r
estadistico
```

```r
# Cuantil de la F
qf(0.95, 4, n - 4)
```

P-value

```r
pf(estadistico, 4, n - 4, lower.tail = F)
```
