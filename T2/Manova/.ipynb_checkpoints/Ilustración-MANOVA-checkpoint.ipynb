```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
from numpy.linalg import inv, det
from MANOVA.manova import MANOVA
```

```python
df = pd.read_csv("data_scores.csv")
df.head()
```

```python
set(df.Method)
```

```python
mean_traditional=df.query("Method=='Traditional'").drop(columns = ['Method']).mean()
```

```python
mean_traditional
```

```python
df.query("Method=='Traditional'").drop(columns = ['Method'])
```

```python
diff_traditional = df.query("Method=='Traditional'").drop(columns = ['Method']) - mean_traditional
```

```python
for x in range(diff_traditional.shape[0]):
    a = diff_traditional.iloc[x].values.reshape(-1,1) @ diff_traditional.iloc[x].values.reshape(-1,1).T
    print(a)
```

```python
# Exploratory Data Analysis (EDA)
print(df.describe(include="all"))
print("-"*50)
print(df.isnull().sum())
print("-"*50)
print(df["Method"].value_counts())
```

```python
# Correlation heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(df.drop(columns = "Method").corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

```python
# Boxplots by method
plt.figure(figsize=(12, 5))
sns.boxplot(x="Method", y="Math_Score", data=df)
plt.title("Boxplot of Math Scores by Teaching Method")
plt.show()
```

```python
# Boxplots by method
plt.figure(figsize=(12, 5))
sns.boxplot(x="Method", y="Science_Score", data=df)
plt.title("Boxplot of Science Scores by Teaching Method")
plt.show()
```

```python
# Running MANOVA
manova_test = MANOVA(df, "Method", ["Math_Score", "Science_Score"])
results = manova_test.fit()
print(results)
```
